```python
import numpy as np

roomData = np.array([1,1,2,2,3,4,4,5,5,5])

priceData = np.array([1000, 5000, 4700, 4100, 8000, 5400, 8900, 9000, 8900, 9900])
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.scatter(roomData, priceData)
plt.xlabel('Variabel X')
plt.ylabel('Variabel Y')
```

```python
from sklearn.linear_model import LinearRegression


roomData = roomData.reshape(-1, 1)
linreg = LinearRegression()
linreg.fit(roomData, priceData)
```

```python
plt.scatter(roomData, priceData)
plt.plot(roomData, linreg.predict(roomData))
```
